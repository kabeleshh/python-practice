```python
letter = input("Enter an alphabet : ")

while not letter.isalpha():
    print("---   Kindly enter an alphabet   ---")
    letter = input("Enter an alphabet : ")

print("Thank you")    
```

```python
num = int(input("Enter a number between 1 to 20: "))

if num<=20:
    while num<=20:
        print(num)
        num+=1
else:
    print("Kindly enter required number")
    
```

```python
user_name = input("Kindly Enter username : ")
password = input("Enter password")

check = True
verified = False

while check:
    if user_name == "Kabelesh" and password == "*0123#":
        print("User verified successfully")
        verified = True
        check = not check
    else:
        print("Invalid Username")
        verified = False
        check = not check

print("Thank You for using our website...")        
    
```
